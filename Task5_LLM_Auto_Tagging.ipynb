```python
from transformers import pipeline
print("Loading model...")
classifier = pipeline("zero-shot-classification", model="typeform/distilbert-base-uncased-mnli")
print("✅ Model ready")
```

```python
tickets = [
    "My computer screen keeps flickering and then goes black",
    "I was charged $49.99 but my plan is only $29.99",
    "How can I change my email address on my account?",
    "When will the new version be released for Mac?",
    "Your customer support is terrible, 5 days no response",
    "The app crashes every time I try to upload a photo",
    "I want to cancel my subscription immediately",
    "Does your product support Windows 11?",
    "I forgot my password and reset link never arrives",
    "The product arrived damaged, box was crushed"
]

categories = ["technical issue", "billing", "account management", "product inquiry", "complaint"]

print(f"✅ {len(tickets)} tickets loaded")
print(f"✅ {len(categories)} categories: {categories}")
```

```python
print("="*60)
print("TASK 5: AUTO TAGGING SUPPORT TICKETS")
print("="*60)

for i, ticket in enumerate(tickets, 1):
    result = classifier(ticket, categories)

    print(f"\n📝 Ticket #{i}: {ticket[:50]}...")
    print(f"   Top 3 tags:")
    print(f"      1. {result['labels'][0].upper()} ({result['scores'][0]:.2%})")
    print(f"      2. {result['labels'][1].upper()} ({result['scores'][1]:.2%})")
    print(f"      3. {result['labels'][2].upper()} ({result['scores'][2]:.2%})")
    print("-" * 40)
```

```python
print("\n" + "="*60)
print("SUMMARY - TAG DISTRIBUTION")
print("="*60)

# Count tags
tag_counts = {}
for ticket in tickets:
    result = classifier(ticket, categories)
    top_tag = result['labels'][0]
    tag_counts[top_tag] = tag_counts.get(top_tag, 0) + 1

for tag, count in tag_counts.items():
    print(f"   {tag.upper()}: {count} tickets ({count/len(tickets)*100:.0f}%)")

print("\n" + "="*60)
print("TASK 5 COMPLETED")
print("="*60)
print("""
✅ Zero-shot classification with LLM
✅ Top 3 tags per ticket
✅ Confidence scores
✅ Categories: Technical, Billing, Account, Product, Complaint
""")
```
